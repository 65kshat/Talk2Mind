```python
## Importing Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import torch
import torch.nn as nn

print("Libraries loaded Successfully.")
```

```python
## Defining Emotion Classes

emotion_classes = ["Anger", "Disgust", "Fear", "Happy", "Neutral", "Sad", "Surprise" ]
print(emotion_classes)
```

```python
## Simulated Visual Emotion Probabilities

visual_probs = np.array([
    0.05,   # Anger
    0.04,   # Disgust
    0.08,   # Fear
    0.55,   # Happy
    0.10,   # Neutral
    0.10,   # Sad
    0.08    # Surprise
])

print("Visual Feature Shape:", visual_probs.shape)
print(visual_probs)
```

```python
## Simulated Speech Emotion Probabilities

speech_probs = np.array([
    0.10,   # Anger
    0.05,   # Disgust
    0.10,   # Fear
    0.40,   # Happy
    0.15,   # Neutral
    0.15,   # Sad
    0.05    # Surprise
])

print("Speech Feature Shape:", speech_probs.shape)
print(speech_probs)
```

```python
## Feature Fusion

fusion_vector = np.concatenate([visual_probs, speech_probs])

print("Fusion Vector Shape: ", fusion_vector.shape)
print(fusion_vector)
```

```python
## Example Questionnaire Scores

depression_score = 18
anxiety_score = 12
stress_score = 20

questionnaire_vector = np.array([depression_score, anxiety_score, stress_score])

print("Questionnaire Shape: ", questionnaire_vector.shape)
print(questionnaire_vector)
```

```python
## Creating a Multimodal Featuer Vector

multimodal_vector = np.concatenate([fusion_vector, questionnaire_vector])

print("Multimodal Shape: ", multimodal_vector.shape)
print(multimodal_vector)
```

```python
## Mental Well-Being Score

visual_positive = (visual_probs[3] + visual_probs[4])
speech_positive = (speech_probs[3] + speech_probs[4])
emotion_score = ((visual_positive * 50) + (speech_positive * 50))

questionnaire_score = max(0, 100 - (depression_score*2 + anxiety_score*1.5 + stress_score*1.5))
mental_wellbeing_score = (0.2*emotion_score + 0.2*emotion_score + 0.6*questionnaire_score)

print(f"Mental Well Being Score: {mental_wellbeing_score:.2f} / 100")
```

```python
## Risk Calculation

if mental_wellbeing_score >= 80:
    category = "Excellent"


elif mental_wellbeing_score >= 65:
    category = "Good"

elif mental_wellbeing_score >= 50:
    category = "Moderate"

elif mental_wellbeing_score >= 35:
    category = "Poor"

else:
    category = "Critical"

print("Category: ", category)
```

```python
from torchvision.models import (efficientnet_b0, EfficientNet_B0_Weights)

## Setting the GPU for models

device = torch.device("cuda")
print("Device: ", device)

if torch.cuda.is_available():
    print("GPU: ", torch.cuda.get_device_name(0))
    print("CUDA Version: ", torch.version.cuda)
else:
    print("CUDA NA")
```

```python
## Loading Visual Model

visual_model = efficientnet_b0(weights = None)
num_features = (visual_model.classifier[1].in_features)
visual_model.classifier[1] = nn.Linear(num_features, 8)

visual_model.load_state_dict(torch.load(r"", map_location = device))

visual_model = visual_model.to(device)
visual_model.eval()
print("Visual Model Loaded.")
```

```python
## Speech BiLSTM Architecture

import torch
import torch.nn as nn

class SpeechBiLSTM(nn.Module):

    def __init__(self, input_size=42, hidden_size=256, num_layers=2, num_classes=8, dropout=0.3):

        super().__init__()

        self.lstm = nn.LSTM(
            input_size=input_size,
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout
        )

        self.dropout = nn.Dropout(dropout)

        self.fc1 = nn.Linear(hidden_size * 2, 256)

        self.relu = nn.ReLU()

        self.fc2 = nn.Linear(256, num_classes)

    def forward(self, x):

        lstm_out, (hidden, cell) = self.lstm(x)

        forward_hidden = hidden[-2]

        backward_hidden = hidden[-1]

        hidden_concat = torch.cat((forward_hidden, backward_hidden), dim=1)

        x = self.dropout(hidden_concat)

        x = self.fc1(x)

        x = self.relu(x)

        x = self.fc2(x)

        return x

print("SpeechBiLSTM Defined")
```

```python
## Audio Model


speech_model = SpeechBiLSTM().to(device)

speech_model.load_state_dict(torch.load(r"", map_location = device))
speech_model = speech_model.to(device)
speech_model.eval()
print("Speech Model Loaded")
```

```python
## Questionnaire Neural Network

class QuestionnaireNN(nn.Module):

    def __init__(self):

        super().__init__()

        self.network = nn.Sequential(

            nn.Linear(30, 64),

            nn.ReLU(),

            nn.Dropout(0.3),

            nn.Linear(64, 32),

            nn.ReLU(),

            nn.Dropout(0.3),

            nn.Linear(32, 5)

        )

    def forward(self, x):

        return self.network(x)

print("QuestionnaireNN Defined")
```

```python
questionnaire_model = QuestionnaireNN()

questionnaire_model.load_state_dict(torch.load(r"", map_location=device))
questionnaire_model = questionnaire_model.to(device)
questionnaire_model.eval()
print("Questionnaire Model Loaded")
```

```python
emotion_labels = ["Neutral", "Calm", "Happy", "Sad", "Angry", "Fearful", "Disgust", "Surprised"]
```

```python
import torch.nn.functional as F

def predict_visual_tensor(image_tensor):

    image_tensor = image_tensor.to(device)

    with torch.no_grad():

        outputs = visual_model(image_tensor)

        probs = F.softmax(outputs, dim=1)

    return probs.cpu().numpy()[0]
```

```python
dummy_image = torch.randn(1, 3, 224, 224).to(device)

visual_probs = predict_visual_tensor(dummy_image)

print(visual_probs)
print("Sum =", visual_probs.sum())
```

```python
def predict_speech_tensor(feature_tensor):

    feature_tensor = feature_tensor.to(device)

    with torch.no_grad():

        outputs = speech_model(feature_tensor)

        probs = F.softmax(outputs, dim=1)

    return probs.cpu().numpy()[0]
```

```python
dummy_audio = torch.randn(1, 150, 42).to(device)

speech_probs = predict_speech_tensor(dummy_audio)

print(speech_probs)
print("Sum =", speech_probs.sum())
```

```python
def predict_questionnaire_tensor(questionnaire_tensor):

    questionnaire_tensor = questionnaire_tensor.to(device)

    with torch.no_grad():

        outputs = questionnaire_model(questionnaire_tensor)

        probs = F.softmax(outputs, dim=1)

    return probs.cpu().numpy()[0]
```

```python
dummy_questionnaire = torch.randn(1, 30).to(device)

questionnaire_probs = predict_questionnaire_tensor(
    dummy_questionnaire
)

print(questionnaire_probs)
print("Sum =", questionnaire_probs.sum())
```
